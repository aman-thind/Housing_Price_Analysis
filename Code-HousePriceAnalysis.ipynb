```python
#filling in the missing values

data.fillna(data.mean(), inplace= True)
```

```python
#defining the outliners observed in the box plot

outliers= ['LotArea','SalePrice','MasVnrArea','BsmtFinSF1','WoodDeckSF','OpenPorchSF','EnclosedPorch','TotalBsmtSF','1stFlrSF','2ndFlrSF','LowQualFinSF','GrLivArea','BsmtHalfBath','BedroomAbvGr','KitchenAbvGr','GarageArea']
```

```python
#cleaninig the outliers

def cleaning_outliers(data, outliers):
    for col in outliers:
        data = data[data[col] < outliers[col]]
    return data
```

```python
#Now that we have fully preprocessed our data, Lets separate our independent variables from the test and train dataset
x = data.iloc[:,:-1].values
y = data.iloc[:,-1].values

# lets split data into train and test format
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3,random_state = 10, shuffle=True)
```

```python
#lets train the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
#Lets now fit the model on the test data set to generate a prediciton
y_pred = regressor.predict(x_test)
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("SalesPrice")
plt.ylabel("Predicted SalesPrice")
plt.show()
```

```python
#lets print prediction
print(y_pred)
```

```python
#lets print accuracy
print("Accuracy --> ", regressor.score(x_test, y_test)*100)
```
